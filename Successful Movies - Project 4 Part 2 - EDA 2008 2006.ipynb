## EDA Code Section

### Imports

```python
import pandas as pd
import numpy as np
```

### Load TMDB Movie Files

```python
tmdbdata_2006 = pd.read_csv('Data/final_tmdb_data_2006.csv.gz')
tmdbdata_2006.head()
```

```python
# drop the first nan row 2000 movies
tmdbdata_2006.drop([0],axis=0, inplace=True)
tmdbdata_2006.head()
```

```python
tmdbdata_2008 = pd.read_csv('Data/final_tmdb_data_2008.csv.gz')
tmdbdata_2008.head()
```

```python
# drop the first nan row 2001 movies
tmdbdata_2008.drop([0],axis=0, inplace=True)
tmdbdata_2008.head()
```

```python
tmdbdata_2006.info()
```

```python
tmdbdata_2008.info()
```

### Combine 2000 & 2001 Movies

```python
## concatenate the tmdb 2000 & 2001 movies
combined_tmdb_20082006 = pd.concat([tmdbdata_2006, tmdbdata_2008],
                      ignore_index=True)
```

```python
combined_tmdb_20082006.info()
```

### 1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
* Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.

```python
filter_financial = (combined_tmdb_20082006['budget'] > 0) | (combined_tmdb_20082006['revenue']>0)
filter_financial
```

```python
filter_financial.value_counts()
```

```python
finpos_movies20082006 = combined_tmdb_20082006.loc[filter_financial==True,:]
finpos_movies20082006
```

```python
finpos_movies20082006.info()
```

### 2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
finpos_movies20082006.groupby('certification')['imdb_id'].count()
```

### 3. What is the average revenue per certification category?

```python
avg_rev20082006 = finpos_movies20082006.groupby('certification')['revenue'].mean()
avg_rev20082006
```

### 4. What is the average budget per certification category?

```python
finpos_movies20082006.groupby('certification')['budget'].mean()
```

## All TMDB Movie Data

```python
title_basics = pd.read_csv('Data/title_basics.csv.gz')
title_basics.head()
```

```python
## example tmdb data combined & basics
merge_tmdbdata_basics = pd.merge(finpos_movies20082006, title_basics, left_on='imdb_id', right_on='tconst', how='inner')
merge_tmdbdata_basics.head()
```

```python
# drop duplicate column merged on
merge_tmdbdata_basics = merge_tmdbdata_basics.drop(columns=['tconst'])
merge_tmdbdata_basics.head()
```

```python
title_ratings = pd.read_csv('Data/title_ratings.csv.gz')
title_ratings.head()
```

```python
## example tmdb data combined & basics & ratings
merge_tmdbbasics_ratings = pd.merge(merge_tmdbdata_basics, title_ratings, left_on='imdb_id', right_on='tconst', how='inner')
merge_tmdbbasics_ratings.head()
```

```python
# drop duplicate column merged on
merge_tmdbbasics_ratings = merge_tmdbbasics_ratings.drop(columns=['tconst'])
merge_tmdbbasics_ratings.head()
```

```python
title_akas = pd.read_csv('Data/title_akas.csv.gz')
title_akas.head()
```

```python
## example tmdb data combined & basics & ratings & akas
merge_tmdbbasicsrate_akas20082006 = pd.merge(merge_tmdbbasics_ratings, title_akas, left_on='imdb_id', right_on='titleId', how='inner')
merge_tmdbbasicsrate_akas20082006.head()
```

```python
# drop duplicate column merged on
merge_tmdbbasicsrate_akas20082006 = merge_tmdbbasicsrate_akas20082006.drop(columns=['titleId'])
merge_tmdbbasicsrate_akas20082006.head()
```

```python
# info on all merged df
merge_tmdbbasicsrate_akas20082006.info()
```

```python
# write merged results to final combined file
merge_tmdbbasicsrate_akas20082006.to_csv("Data/tmdb_results_combined20082006.csv.gz", compression="gzip", index=False)
```
